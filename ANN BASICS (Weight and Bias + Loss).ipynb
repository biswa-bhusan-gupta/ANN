```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
Obv=1000
```

```python
x=np.random.uniform(low=-10,high=10,size=(Obv,1))
z=np.random.uniform(low=-10,high=10,size=(Obv,1))
```

```python
Input=np.column_stack((x,z))
```

```python
Input.shape
```

```python
Input
```

```python
Noise=np.random.uniform(low=-1,high=1,size=(Obv,1))
```

```python
Noise
```

```python
Target= 2*x-3*z+5+Noise
```

```python
Target
```

```python
Target.shape
```

```python
Range=0.1
```

```python
Weight=np.random.uniform(-Range,Range,(2,1))
Bias=np.random.uniform(-Range,Range,1)
```

```python
Weight
```

```python
Bias
```

```python
Learning_Rate=0.02
```

```python
for i in range (100):
    Output=np.dot(Input,Weight)+Bias
    Delta=Output-Target
    
    Loss=np.sum(Delta ** 2)/2/Obv
    print(Loss)
    
    Delta_Scaled=Delta/Obv
    
    Weight=Weight-Learning_Rate * np.dot(Input.T,Delta_Scaled)
    Bias=Bias-Learning_Rate * np.sum(Delta_Scaled)
```

```python
print(Weight,Bias)
```

```python
plt.plot(Output,Target)
```
