```python
Geo=pd.get_dummies(x['Geography'],drop_first=True)
```

```python
x=pd.concat([x,Geo,Gen],axis=1)
```

```python
x=x.drop(['Geography','Gender'],axis=1)
x
```

```python
X=x.values
Y=y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization, Dropout
from tensorflow.keras.activations import relu, sigmoid
```

```python
def Create_Model(layers,activation):
    Model=Sequential()
    for i,nodes in enumerate(layers):
        if i==0:
            Model.add(Dense(nodes,input_dim=x_train.shape[1],activation=activation))
            Model.add(Dropout(0.3))
        else:
            Model.add(Dense(nodes,activation=activation))
            Model.add(Dropout(0.3))
            
    Model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
    Model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return Model

KC=KerasClassifier(build_fn=Create_Model,verbose=2)

activation=['relu','sigmoid']
layers=[(20,),(40,20)]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Grid=dict(layers=layers,activation=activation,batch_size=[128,256],epochs=[30])
```

```python
GSC=GridSearchCV(estimator=KC,param_grid=Grid,cv=5,verbose=2)
GSC.fit(x_train,y_train)
```
