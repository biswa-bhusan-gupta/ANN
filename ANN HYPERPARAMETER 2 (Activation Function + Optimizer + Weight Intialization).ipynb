```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Data=pd.read_csv('Cancer_Classification.csv')
```

```python
Data
```

```python
Data.columns.unique()
```

```python
x = Data.drop('benign_0__mal_1',axis=1).values
y = Data['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
```

```python
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Embedding, Flatten, LeakyReLU, BatchNormalization, Dropout
from tensorflow.keras.activations import relu, sigmoid
```

```python
# Number of Epochs and Batch Size:
```

```python
def Create_Model1():
    Model1=Sequential()
    Model1.add(Dense(units=12,kernel_initializer='glorot_uniform',activation='sigmoid',input_dim=30))
    Model1.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
    Model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return Model1

KC1=KerasClassifier(build_fn=Create_Model1,verbose=2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Grid1={
      'batch_size':[10,20,40,80,100],
      'epochs':[10,50,100]
     }
```

```python
GSC1=GridSearchCV(estimator=KC1,param_grid=Grid1,cv=5,verbose=2)
GSC1.fit(x_train,y_train)
```

```python
GSC1.best_params_
```

```python
GSC1.best_score_
```

```python
# Optimizer:
```

```python
def Create_Model2(optimizer='adam'):
    Model2=Sequential()
    Model2.add(Dense(units=12,kernel_initializer='glorot_uniform',activation='sigmoid',input_dim=30))
    Model2.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
    Model2.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return Model2

KC2=KerasClassifier(build_fn=Create_Model2,epochs=100,batch_size=10,verbose=2)
```

```python
optimizer = ['rmsprop','adagrad','adamax','adam']
Grid2=dict(optimizer=optimizer)
```

```python
GSC2=GridSearchCV(estimator=KC2,param_grid=Grid2,cv=5,verbose=2)
GSC2.fit(x_train,y_train)
```

```python
GSC2.best_params_
```

```python
GSC2.best_score_
```

```python
def Create_Model4(activation):
    Model4=Sequential()
    Model4.add(Dense(units=12,kernel_initializer='he_uniform',activation=activation,input_dim=30))
    Model4.add(Dense(units=1,kernel_initializer='he_uniform',activation='sigmoid'))
    Model4.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return Model4

KC4=KerasClassifier(build_fn=Create_Model4,verbose=2,epochs=100,batch_size=10)
```

```python
# Learning Rate and Momentum:
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
def Create_Model5(learn_rate=0.01,momentum=0):
    Model5=Sequential()
    Model5.add(Dense(units=12,kernel_initializer='he_uniform',activation='tanh',input_dim=30))
    Model5.add(Dense(units=1,kernel_initializer='he_uniform',activation='sigmoid'))
    
    optimizer=SGD(learning_rate=learn_rate,momentum=momentum,name='SGD')
    Model5.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return Model5

KC5=KerasClassifier(build_fn=Create_Model5,verbose=2,epochs=100,batch_size=10)
```

```python
learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
momentum = [0.0, 0.2, 0.4, 0.6, 0.8, 0.9]
Grid5=dict(learn_rate=learn_rate,momentum=momentum)
```

```python
GSC5=GridSearchCV(estimator=KC5,param_grid=Grid5,cv=5,verbose=2)
GSC5.fit(x_train,y_train)
```

```python
GSC5.best_params_
```

```python
GSC5.best_score_
```
