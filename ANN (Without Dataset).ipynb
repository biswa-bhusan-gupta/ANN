```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Obv=100
```

```python
x=np.random.uniform(low=-10,high=10,size=(Obv,1))
z=np.random.uniform(low=-10,high=10,size=(Obv,1))
```

```python
x
```

```python
Input=np.column_stack((x,z))
```

```python
Input
```

```python
Noise=np.random.uniform(low=-1,high=1,size=(Obv,1))
```

```python
Target= 2*x-3*z+5+Noise
```

```python
Target
```

```python
np.savez('TF_INTRO',Inputs=Input,Targets=Target)
```

```python
Training_Data=np.load('TF_INTRO.npz')
```

```python
No_of_Inputs=2
No_of_Outputs=1
```

```python
Model=tf.keras.Sequential([
                            tf.keras.layers.Dense(No_of_Outputs)
                          ])
Model.compile(optimizer='SGD',loss='mean_squared_error')
```

```python
Model.fit(Training_Data['Inputs'],Training_Data['Targets'],epochs=100,verbose=0)
```

```python
Model.fit(Training_Data['Inputs'],Training_Data['Targets'],epochs=100,verbose=1)
```

```python
Model.fit(Training_Data['Inputs'],Training_Data['Targets'],epochs=100,verbose=2)
```
